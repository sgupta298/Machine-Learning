```python
import matplotlib.pyplot as plt
x=[2,5,7]
y=[4,6,23]
x1=[45,3,1]
y1=[96,34,65]
plt.xlabel("x")
plt.ylabel("y")
plt.scatter(x,y,label='first',marker='x',s=245,color='r')
plt.plot(x1,y1,label='second')
plt.legend()
plt.grid(color='grey')
plt.show()
```
